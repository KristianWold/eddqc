# Marginals, Plotting

```python
import sys
sys.path.insert(0, '../../src/')

import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf


#np.set_printoptions(threshold=sys.maxsize)
np.set_printoptions(precision=4)

import os
os.environ["CUDA_VISIBLE_DEVICES"] = "-1"
tf.compat.v1.logging.set_verbosity(tf.compat.v1.logging.ERROR)
```

## Radial, L = 5

```python
[num_bins, FF_hist] = np.loadtxt('bookkeeping/figure3/radial_FF_theory.txt').T
[_, AI_hist] = np.loadtxt('bookkeeping/figure3/radial_AI_theory.txt').T
[exp_bins, exp_hist] = np.loadtxt("bookkeeping/figure3/radial_L=5.txt").T
bootstrap_err = np.loadtxt('bookkeeping/figure3/radial_std_L=5.txt')
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(num_bins, FF_hist, 'b-', 
        label='Theoretic, Integrable',
        )

ax.plot(num_bins, AI_hist, 'g-', 
         label='Theoretic, Non-Integrable',
         )

ax.fill_between(exp_bins, (exp_hist-bootstrap_err), (exp_hist+bootstrap_err), alpha=0.25, color="coral" )
ax.plot(exp_bins, exp_hist, 'o-', label='Experimental, T = 5', color="coral")


plt.xlim(0, 1)
plt.ylim(-0.1, 3.)
plt.xlabel(r'Radius ($r$)')
plt.ylabel(r'$P(r)$')
#ax.grid(True)
#ax.set_yticklabels([])
plt.legend()
plt.savefig("figures/transition_radial_L=5.svg", bbox_inches='tight')

plt.show()
```

## Angular, L = 5

```python
[num_bins, FF_hist] = np.loadtxt('bookkeeping/figure3/angular_FF_theory.txt').T
[_, AI_hist] = np.loadtxt('bookkeeping/figure3/angular_AI_theory.txt').T

[exp_bins, exp_hist] = np.loadtxt("bookkeeping/figure3/angular_L=5.txt").T
bootstrap_err = np.loadtxt('bookkeeping/figure3/angular_std_L=5.txt')

plt.rc('font', size=16) 

fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(num_bins, FF_hist, 'b-', 
#        label='Theoretic, Integrable',
        )

ax.plot(num_bins, AI_hist, 'g-', 
#         label='Theoretic, Non-Integrable',
         )

ax.fill_between(exp_bins, (exp_hist-bootstrap_err), (exp_hist+bootstrap_err), alpha=0.25, color="coral" )
ax.plot(exp_bins, exp_hist, 'o-', label='Experimental, T = 5', color="coral")


plt.xlim(-np.pi, np.pi)
plt.ylim(0.03, 0.36)
plt.xlabel(r'Angle ($\theta$)')
plt.ylabel(r'$P(\theta)$')
plt.legend()
plt.savefig("figures/transition_angular_L=5.svg", bbox_inches='tight')

plt.show()
```
