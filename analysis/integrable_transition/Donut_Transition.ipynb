# Donut Transition

```python
import sys
sys.path.insert(0, '../../src/')

import numpy as np
import matplotlib.pyplot as plt
import pickle as pkl
import tensorflow as tf

from spectrum import channel_spectrum, complex_spacing_ratio, Spectrum, coat_spectrum


#np.set_printoptions(threshold=sys.maxsize)
np.set_printoptions(precision=4)

import os
os.environ["CUDA_VISIBLE_DEVICES"] = "-1"
tf.compat.v1.logging.set_verbosity(tf.compat.v1.logging.ERROR)
```

```python
def model_to_csr(model_list):
    spectrum_list = []
    csr_integrable_list = []
    for model in model_list:
        channel = model.channel
        spectrum = channel_spectrum(channel)

        csr = complex_spacing_ratio(spectrum)

        spectrum_list.append(spectrum)
        csr_integrable_list.append(csr)

    return csr_integrable_list
```

```python
c_max = 0.6

def plot_csr(csr_coat, plotname, L, turn_off_yticks=False):
    N = csr_coat.shape[0]

    ticks = np.linspace(0, N-1, 5)
    labels = np.linspace(-1, 1, 5)

    fig = plt.figure(figsize=(5, 5))
    fig.tight_layout()
    plt.imshow(csr_coat)
    #plt.clim(0, c_max)
    plt.colorbar()


    plt.xlabel("Re z")
    plt.xticks(ticks, labels=labels)

    if turn_off_yticks:
        plt.yticks([])
    else:
        plt.ylabel("Im z")
        plt.yticks(ticks, labels=labels)

    # put label with "L=" in the top left corner 
    plt.text(0.05, 0.95, "T=" + str(L), transform=plt.gca().transAxes, fontsize=24, verticalalignment='top', color='white')
    plt.savefig(plotname, bbox_inches='tight')
    plt.show()
```

```python
#from matplotlib import rc
#rc('font',**{'family':'serif','serif':['Times']})
#rc('text', usetex=True)

# Increase font size
plt.rcParams.update({'font.size': 14})
```

## Retrieve CSR

### L = 5

```python
model_list = pkl.load(open(f'../fitting_maps/models/integrable_model_4_L=5.model', 'rb'))

csr_integrable_list = model_to_csr(model_list)
csr_integrable = Spectrum(csr_integrable_list)

csr_integrable_coat = coat_spectrum(csr_integrable, sigma = 0.05, grid_size=200)

norm = np.sum(csr_integrable_coat)/10000
csr_integrable_coat /= norm

filename = "bookkeeping/figure3/csr_raw_L=5.txt"
np.savetxt(filename, np.real(csr_integrable.get_spectrum()))

filename = "bookkeeping/figure3/csr_heatmap_blurred_L=5.txt"
np.savetxt(filename, csr_integrable_coat)


plot_csr(csr_integrable_coat, 
        'figures/integrable_crs_retrieved_L=5.svg', 
        L=5, 
        turn_off_yticks=False)
```
